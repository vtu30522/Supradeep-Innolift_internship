Import Libraries

```
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    classification_report
)
```

Load Dataset

```
df = pd.read_csv("/content/drive/MyDrive/track_2/creditcard.csv")

print("Dataset Loaded Successfully")
display(df.head())

print("\nDataset Shape:")
print(df.shape)
```

Data Preprocessing

Check Missing Values

```
print("Missing Values")

display(df.isnull().sum())
df.dropna(inplace=True)

print("Dataset Shape After Cleaning")
print(df.shape)
```

Split Features and Target

```
X = df.drop("Class", axis=1)

y = df["Class"]

print("Features Shape:", X.shape)
print("Target Shape:", y.shape)
```

Train-Test Split

```
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Samples:", len(X_train))
print("Testing Samples:", len(X_test))
```

Train Random Forest Model

```
rf_model = RandomForestClassifier(
    n_estimators=300,
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)

print("Random Forest Model Trained Successfully")
```

Make Predictions

```
y_pred = rf_model.predict(X_test)

print("Prediction Completed")
```

Evaluate Accuracy

```
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)

plt.figure(figsize=(4,4))
plt.bar(['Accuracy'], [accuracy], color='blue')
plt.ylim(0,1)
plt.title('Accuracy Score')
plt.ylabel('Score')
plt.show()
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, y_pred)

print("="*50)
print("ACCURACY EVALUATION REPORT")
print("="*50)

print(f"Accuracy Score : {accuracy:.4f}")

if accuracy >= 0.95:
    print("Evaluation : Excellent Accuracy")
elif accuracy >= 0.85:
    print("Evaluation : Good Accuracy")
else:
    print("Evaluation : Needs Improvement")
```

Evaluate Precision

```
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_pred)

print("="*50)
print("PRECISION EVALUATION REPORT")
print("="*50)

print(f"Precision Score : {precision:.4f}")

if precision >= 0.95:
    print("Evaluation : Very Few False Positives")
elif precision >= 0.85:
    print("Evaluation : Good Precision")
else:
    print("Evaluation : Precision Needs Improvement")
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_pred)

print("Precision:", precision)

plt.figure(figsize=(4,4))
plt.bar(['Precision'], [precision], color='green')
plt.ylim(0,1)
plt.title('Precision Score')
plt.ylabel('Score')
plt.show()
```

Evaluate Recall

```
from sklearn.metrics import recall_score

recall = recall_score(y_test, y_pred)

print("Recall:", recall)

plt.figure(figsize=(4,4))
plt.bar(['Recall'], [recall], color='orange')
plt.ylim(0,1)
plt.title('Recall Score')
plt.ylabel('Score')
plt.show()
from sklearn.metrics import recall_score

recall = recall_score(y_test, y_pred)

print("="*50)
print("RECALL EVALUATION REPORT")
print("="*50)

print(f"Recall Score : {recall:.4f}")

if recall >= 0.95:
    print("Evaluation : Excellent Fraud Detection")
elif recall >= 0.85:
    print("Evaluation : Good Fraud Detection")
else:
    print("Evaluation : Many Fraud Cases Missed")
```

Evaluate f1 score

```
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)

print("F1 Score:", f1)

plt.figure(figsize=(4,4))
plt.bar(['F1 Score'], [f1], color='red')
plt.ylim(0,1)
plt.title('F1 Score')
plt.ylabel('Score')
plt.show()
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)

print("="*50)
print("F1 SCORE EVALUATION REPORT")
print("="*50)

print(f"F1 Score : {f1:.4f}")

if f1 >= 0.95:
    print("Evaluation : Excellent Balance Between Precision and Recall")
elif f1 >= 0.85:
    print("Evaluation : Good Overall Performance")
else:
    print("Evaluation : Model Needs Further Optimization")
```

Evaluate Confusion matrix

```
from sklearn.metrics import ConfusionMatrixDisplay

fig, ax = plt.subplots(figsize=(6,6))

ConfusionMatrixDisplay.from_estimator(
    rf_model,
    X_test,
    y_test,
    cmap=plt.cm.Blues,
    ax=ax
)

ax.set_title("Confusion Matrix")

plt.show()
from sklearn.metrics import classification_report

print("="*50)
print("CLASSIFICATION REPORT")
print("="*50)

print(classification_report(y_test, y_pred))
```

Test Sample Predictions

```
sample_transactions = X_test.iloc[:5]

predictions = rf_model.predict(sample_transactions)

print("Predictions")

for i, pred in enumerate(predictions):
    if pred == 1:
        print(f"Transaction {i+1}: Fraud")
    else:
        print(f"Transaction {i+1}: Genuine")
```
